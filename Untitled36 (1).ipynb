```python
hourly_demand = bike_df.groupby(
    pd.Grouper(
        key="start_time",
        freq="H"
    )
).size().reset_index(name="trip_count")

hourly_demand.head()
```

```python
hourly_demand.shape
```

```python
weather_hourly = climate_df[
    [
        "date_time_lst",
        "temp_°c",
        "rel_hum_%",
        "wind_spd_km_h",
        "visibility_km",
        "stn_press_kpa"
    ]
].copy()

weather_hourly = weather_hourly.drop_duplicates(subset=["date_time_lst"])
```

```python
merged_df = pd.merge(
    hourly_demand,
    weather_hourly,
    left_on="start_time",
    right_on="date_time_lst",
    how="inner"
)

print("Merged shape:", merged_df.shape)
display(merged_df.head())
```

```python
merged_df.isnull().sum()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=merged_df,
    x="temp_°c",
    y="trip_count"
)

plt.title("Temperature vs Bike Demand")
plt.xlabel("Temperature (°C)")
plt.ylabel("Hourly Trip Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=merged_df,
    x="rel_hum_%",
    y="trip_count"
)

plt.title("Humidity vs Bike Demand")
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Hourly Trip Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=merged_df,
    x="wind_spd_km_h",
    y="trip_count"
)

plt.title("Wind Speed vs Bike Demand")
plt.xlabel("Wind Speed")
plt.ylabel("Hourly Trip Count")
plt.tight_layout()
plt.show()
```

```python
corr_cols = [
    "trip_count",
    "temp_°c",
    "rel_hum_%",
    "wind_spd_km_h",
    "visibility_km",
    "stn_press_kpa"
]

plt.figure(figsize=(8,6))

sns.heatmap(
    merged_df[corr_cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Between Weather Variables and Bike Demand")
plt.tight_layout()
plt.show()
```

```python
pickups = bike_df.groupby("start_station_name").size()
returns = bike_df.groupby("end_station_name").size()

station_balance = pd.DataFrame({
    "pickups": pickups,
    "returns": returns
}).fillna(0)

station_balance["imbalance"] = station_balance["pickups"] - station_balance["returns"]

station_balance["total_activity"] = station_balance["pickups"] + station_balance["returns"]

station_balance.head()
```

```python
empty_risk = station_balance.sort_values(
    "imbalance",
    ascending=False
).head()

empty_risk
```

```python
plt.figure(figsize=(12,6))

empty_risk["imbalance"].plot(kind="bar")

plt.title("Top Empty-Risk Stations")
plt.xlabel("Station")
plt.ylabel("Pickups - Returns")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
full_risk = station_balance.sort_values(
    "imbalance",
    ascending=True
).head()

full_risk
```

```python
plt.figure(figsize=(12,6))

full_risk["imbalance"].plot(kind="bar")

plt.title("Top Full-Risk Stations")
plt.xlabel("Station")
plt.ylabel("Pickups - Returns")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
merged_df.to_csv("bike_weather_merged_hourly.csv", index=False)

station_balance.to_csv("station_reliability_balance.csv")

hourly_demand.to_csv("hourly_bike_demand.csv", index=False)

print("Files saved successfully.")
```
